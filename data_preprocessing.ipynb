```python
'''Importing Necessary Libraries'''
import glob
import os
import pandas as pd
import geopandas as gpd
```

```python
'''Loading Street Crime Files'''

# loading all street crime CSV files
crime_files = glob.glob("crime_data/*/*street*.csv")

crime_list = []

# Combining all monthly crime files into one dataframe

for file in crime_files:
    df = pd.read_csv(file)
    crime_list.append(df)

crime_df = pd.concat(crime_list, ignore_index=True)
```

```python
'''Loading Outcomes Files'''

# loading all outcomes CSV files
outcome_files = glob.glob("crime_data/*/*outcomes*.csv")

outcome_list = []

# Combining all monthly outcome files into one dataframe
for file in outcome_files:
    df = pd.read_csv(file)
    outcome_list.append(df)

outcomes_df = pd.concat(outcome_list, ignore_index=True)
```

```python
# 7. DATA CLEANING
```

```python
# creating flags for analysis
merged_df["has_crime_id"] = merged_df["Crime ID"].notnull()

# adding row-level coordinate validity flags
merged_df["has_valid_coords_crime"] = (
    merged_df["Longitude_crime"].notnull() |
    merged_df["Latitude_crime"].notnull()
)

merged_df["has_valid_lsoa_for_mapping_crime"] = (
    merged_df["Longitude_crime"].notnull() &
    merged_df["Latitude_crime"].notnull() &
    merged_df["LSOA name_crime"].notnull() &
    merged_df["LSOA code_crime"].notnull()
)

merged_df["has_valid_coords_outcome"] = (
    merged_df["Longitude_outcome"].notnull() |
    merged_df["Latitude_outcome"].notnull()
)

merged_df["has_valid_lsoa_for_mapping_outcome"] = (
    merged_df["Longitude_outcome"].notnull() &
    merged_df["Latitude_outcome"].notnull() &
    merged_df["LSOA name_outcome"].notnull() &
    merged_df["LSOA code_outcome"].notnull()
)

merged_df["has_outcome"] = merged_df["Outcome type"].isnull() #outcome indicator column
```

```python
'''Clean Crime ID column created'''
merged_df["crime_id_clean"] = pd.factorize(merged_df["Crime ID"])[0] + 1

'''How many times each Crime ID appears'''
merged_df["crime_id_count"] = merged_df.groupby("Crime ID")["Crime ID"].transform("count")
```

```python
'''Coordinate range'''
merged_df = merged_df[
    (merged_df["Latitude_crime"].between(50, 52.5)) &
    (merged_df["Longitude_crime"].between(-2, 1))
]
```

```python
'''Date Formatting for Month'''

# converting the `Month` columns into datetime format so time-based analysis can be performed
merged_df.loc[:, "Month_crime"] = pd.to_datetime(merged_df["Month_crime"])
merged_df.loc[:, "Month_outcome"] = pd.to_datetime(merged_df["Month_outcome"])
```

```python
merged_df
```

```python
'''Preparing Map Data - Crime Points'''

# Create geometry in original coordinate system (WGS84)
merged_df.loc[:, "geometry_wgs84"] = gpd.points_from_xy(
    merged_df["Longitude_crime"],
    merged_df["Latitude_crime"],
    crs="EPSG:4326"
)

# Convert to a GeoSeries and reproject to Web Mercator for basemap compatibility
geometry_3857 = gpd.GeoSeries(merged_df["geometry_wgs84"], crs="EPSG:4326").to_crs(epsg=3857)

# Store reprojected geometry back into merged_df
merged_df.loc[:, "geometry_3857"] = geometry_3857

merged_df
```

```python
'''Preparing map data: LSOA Hotspots'''

# creating LSOA-level aggregated hotspot metrics

lsoa_summary_crime = (
    merged_df[merged_df["has_valid_lsoa_for_mapping_crime"]]
    .groupby(["LSOA code_crime", "LSOA name_crime", "Falls within_crime"], as_index=False)
    .agg(
        lsoa_crime_count=("Crime ID", "size"),
        lsoa_mean_longitude_crime=("Longitude_crime", "mean"),
        lsoa_mean_latitude_crime=("Latitude_crime", "mean")
    )
)

# merging LSOA-level metrics back into merged_df

merged_df = merged_df.merge(
    lsoa_summary_crime,
    on=["LSOA code_crime", "LSOA name_crime", "Falls within_crime"],
    how="left"
)

merged_df
```
